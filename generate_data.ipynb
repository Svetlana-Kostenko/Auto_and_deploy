```python
import csv
import random
import string
from datetime import datetime
```

```python
# Список категорий товаров и их наименований
categories = {
    "бытовая химия": [
 "Средство для мытья посуды универсальное",
 "Стиральный порошок для цветного белья",
 "Чистящий крем для сантехники",
 "Средство для мытья окон без разводов",
 "Порошок для прочистки труб"
],
    "посуда": [
 "Набор тарелок глубоких 6 шт",
 "Кастрюля нержавеющая 3 л",
 "Набор столовых приборов на 6 персон",
 "Сковорода с антипригарным покрытием 26 см",
 "Чайный сервиз на 6 персон"
]
 ,
    "продукты питания": [
 "Молоко пастеризованное 2.5%",
 "Куриные яйца С1",
 "Хлеб цельнозерновой",
 "Картофель очищенный",
 "Бананы свежие"
],
    "средства гигиены" :  [
    "Зубная паста отбеливающая",
    "Гель для душа увлажняющий",
    "Дезодорант-антиперспирант",
    "Шампунь для нормальных волос",
    "Ватные диски косметические"
],
    "товары для животных": [
 "Лежанка для животных с бортиками",
 "Миски устойчивые для корма и воды",
 "Лоток с сеткой для кошек",
 "Щетка-пуходерка для вычесывания",
 "Влажные салфетки для гигиены животных"
]
}


def generate_doc_id():
    # Генерация уникального идентификатора чека
    timestamp = datetime.now().strftime("%Y%m%d%H%M%S")
    random_suffix = ''.join(random.choices(string.ascii_uppercase + string.digits, k=4))
    return f"{timestamp}-{random_suffix}"

def generate_random_price():
    # Генерация случайной цены
    return round(random.uniform(50, 5000), 2)

def generate_random_discount(price):
    # Генерация случайной скидки
    discount_rate = random.uniform(0, 0.3)  # скидка до 30%
    return round(discount_rate, 2)

def generate_receipt(apteka, kassa):
    #  Генерация данных чека
    doc_id = generate_doc_id()
    category = random.choice(list(categories.keys()))
    item = random.choice(categories[category])
    amount = random.randint(1, 5)
    price = generate_random_price()
    discount = generate_random_discount(price)
    
    return {
        "doc_id": apteka + '-' + kassa + '_' + doc_id,
        "item": item,
        "category": category,
        "amount": amount,
        "price": price,
        "discount": discount
    }

def generate_multiple_receipts(num_receipts, apteka, kassa):
    # Генерация нескольких чеков
    receipts = []
    for _ in range(num_receipts):
        receipts.append(generate_receipt(apteka, kassa))
    return receipts

# пусть аптеки имеют номера от 10 до 19 вкл
pharmacy = list(range(10, 20))
n = random.randint(1, len(pharmacy))
# сегодня работают вот эти аптеки
pharmacy_today = random.sample(pharmacy, n)
# в каждой аптеке 4 кассы
cash_register = list(range(1, 5))


for item in pharmacy_today:
    n = random.randint(1, len(cash_register))
    # кассы, которые рбаотают сегодня в конкретной аптеке
    n_regs = random.sample(cash_register, n)
    for reg in n_regs:
        сnt = random.randint(10, 100)
        receipts = generate_multiple_receipts(сnt, str(item), str(reg))
        # запись в файл
        with open(f'data/{str(item)}-{str(reg)}.csv', 'w', newline='', encoding='utf-8') as file:
            fieldnames = ["doc_id","item","category","amount","price","discount"]
            writer = csv.DictWriter(file, fieldnames=fieldnames)            
            writer.writeheader()
            writer.writerows(receipts)
```
